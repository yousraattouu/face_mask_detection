```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

num_of_classes = 2

model = Sequential()

# Couches de convolution et de pooling
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(128, 128, 3)))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Couche de flattening pour convertir les features en un vecteur 1D
model.add(Flatten())

# Couches fully connected (dense)
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))

# Couche de sortie avec activation 'sigmoid' pour la classification binaire
model.add(Dense(num_of_classes, activation='sigmoid'))

# Afficher le résumé du modèle
model.summary()
```

```python
num_of_classes =

model = keras.Sequential()

model.add(keras.layers.Conv2D(32, kernel_size=(3,3), activation='relu', input_shape=(128,128,3)))#La première couche convolutive avec 32 filtres,
#une taille de noyau de (3,3), une fonction d'activation 'relu', et une forme d'entrée de (128,128,3).
#Cette couche est responsable de détecter des caractéristiques dans l'image.
model.add(keras.layers.MaxPooling2D(pool_size=(2,2)))#La première couche de sous-échantillonnage (pooling) avec une taille de pool de (2,2).
# Elle réduit les dimensions spatiales de la représentation de l'image.


model.add(keras.layers.Conv2D(64, kernel_size=(3,3), activation='relu'))#Une deuxième couche convolutive avec 64 filtres et une taille de noyau de (3,3)
                                                                        # suivie d'une autre couche de sous-échantillonnage.
model.add(keras.layers.MaxPooling2D(pool_size=(2,2)))#Une deuxième couche de sous-échantillonnage avec une taille de pool de (2,2).

model.add(keras.layers.Flatten())#Cette couche transforme la sortie des couches précédentes en un vecteur unidimensionnel,
                                 # préparant les données pour les couches complètement connectées.

model.add(keras.layers.Dense(128, activation='relu'))#Une couche entièrement connectée (dense) avec 128 neurones et une fonction d'activation 'relu'
                                                     #Elle cherche à apprendre des relations plus complexes dans les données.
model.add(keras.layers.Dropout(0.5))# Une couche de régularisation qui désactive aléatoirement 50% des neurones pendant l'entraînement.
                                    # Cela aide à prévenir le surajustement (overfitting).

model.add(keras.layers.Dense(64, activation='relu'))#Une deuxième couche entièrement connectée avec 64 neurones et une fonction d'activation 'relu'
                                                    #, suivie d'une autre couche de dropout.
model.add(keras.layers.Dropout(0.5))#Une autre couche de régularisation.


model.add(keras.layers.Dense(num_of_classes, activation='sigmoid'))#La couche de sortie avec le nombre de neurones égal au nombre de classes
             #(2 dans votre cas) et une fonction d'activation 'sigmoid'. Cela convient pour un problème de classification binaire.
```

```python
# compile the neural network
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(x_train_scaled, y_train, validation_split=0.1, epochs=5)
```

```python
loss, accuracy = model.evaluate(x_test_scaled, y_test)
print('Test Accuracy =', accuracy)
```

```python
h = history

# plot the loss value
plt.plot(h.history['loss'], label='train loss')
plt.plot(h.history['val_loss'], label='validation loss')
plt.legend()
plt.show()

# plot the accuracy value
plt.plot(h.history['accuracy'], label='train accuracy')
plt.plot(h.history['val_accuracy'], label='validation accuracy')
plt.legend()
plt.show()
```
